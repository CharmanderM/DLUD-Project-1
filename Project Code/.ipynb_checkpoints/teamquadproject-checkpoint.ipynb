```python
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2

DATADIR = "../project_dataset"
DATADIR2 = "../validation_dataset"
CATEGORIES = ["motherbaby","pornstar"]

for category in CATEGORIES:
    path = os.path.join(DATADIR, category) #path to motherbaby and pornstar dir
    for img in os.listdir(path):
        img_array = cv2.imread(os.path.join(path,img), cv2.IMREAD_GRAYSCALE)
        # img_array = cv2.imread(os.path.join(path,img))
        plt.imshow(img_array, cmap="gray")
        plt.show()
        break
    break
```

```python
print(img_array.shape)
```

```python
IMG_SIZE = 100

new_array = cv2.resize(img_array, (IMG_SIZE, IMG_SIZE))
plt.imshow(new_array, cmap = 'gray')
plt.show()
```

```python
training_data = []

def creat_training_data():
    for category in CATEGORIES:
        path = os.path.join(DATADIR, category) #path to motherbaby and pornstar dir
        class_num = CATEGORIES.index(category)
        for img in os.listdir(path):
#             try:
            img_array = cv2.imread(os.path.join(path,img), cv2.IMREAD_GRAYSCALE)
            if img_array is not None:
                new_array = cv2.resize(img_array[0], (IMG_SIZE, IMG_SIZE))
                training_data.append([new_array, class_num])
            else:
                pass
            
#             except Exception as e:
#                 pass
creat_training_data()
```

```python
import random

random.shuffle(training_data)
```

```python
help(training_data)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D, Flatten, Dropout, MaxPooling2D
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import os
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
batch_size=128
epochs=15
```

```python
image_generator=ImageDataGenerator(rescale=1./255)
```

```python
train_data_gen=image_generator.flow_from_directory(directory=DATADIR,batch_size=batch_size,
                                                   shuffle=True,target_size=(IMG_SIZE,IMG_SIZE),class_mode='binary')
```

```python
sample_training_images, _ = next(train_data_gen)
def plotImages(images_arr):
    fig, axes = plt.subplots(1, 5, figsize=(20,20))
    axes = axes.flatten()
    for img, ax in zip( images_arr, axes):
        ax.imshow(img)
        ax.axis('off') 
    plt.tight_layout()
    plt.show()
plotImages(sample_training_images[:5])
```

```python
validation_img_generator=ImageDataGenerator(rescale=1./255)
val_data_gen = validation_img_generator.flow_from_directory(directory=DATADIR2,
                                                            batch_size=batch_size,shuffle=True,target_size=(IMG_SIZE,IMG_SIZE),class_mode='binary')
```

```python
sample_training_images, _ = next(val_data_gen)
def plotImages(images_arr):
    fig, axes = plt.subplots(1, 5, figsize=(20,20))
    axes = axes.flatten()
    for img, ax in zip( images_arr, axes):
        ax.imshow(img)
        ax.axis('off') 
    plt.tight_layout()
    plt.show()
plotImages(sample_training_images[:5])
```

```python
model = Sequential([
Conv2D(16, 3, padding='same', activation='relu', input_shape=(IMG_SIZE, IMG_SIZE ,3)), 
    MaxPooling2D(),
Conv2D(32, 3, padding='same', activation='relu'),
MaxPooling2D(),
Conv2D(64, 3, padding='same', activation='relu'),
MaxPooling2D(),
Flatten(),
Dense(512, activation='relu'),
Dense(1)
])
model.compile(optimizer='adam',loss=tf.keras.losses.BinaryCrossentropy(from_logits=True),metrics=['accuracy'])
model.summary()
 
```

```python
help(Conv2D)
```

```python
history = model.fit_generator(train_data_gen,steps_per_epoch = 1202//batch_size,
                              epochs=epochs,validation_data=val_data_gen,validation_steps=1202//batch_size)
```

```python
acc=history.history['accuracy']
```

```python
val_acc = history.history['val_accuracy']
```

```python
loss = history.history['loss']
val_loss = history.history['val_loss']
```

```python
epochs_range = range(epochs)
```

```python
plt.subplot(1,2,1)
plt.plot(epochs_range,acc,label="Training Accuracy")
plt.plot(epochs_range,val_acc,label="Validation Accuracy")
plt.legend(loc="lower right")
plt.title("Training and Validation Accuracy")
plt.subplot(1,2,2)
plt.plot(epochs_range,loss,label='Trainig Loss')
plt.plot(epochs_range,val_loss,label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.show()
```
